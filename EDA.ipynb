# Car Sales Advertisement Analysis

## Introduction
This project analyzes car sales advertisements. The goal is to explore the distribution of car prices, model years, odometer readings, and other features to gain insights into the dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('vehicles_us.csv')

# Fill missing values
df['model_year'] = df.groupby('model')['model_year'].transform(lambda x: x.fillna(x.median()))
df['cylinders'] = df.groupby('model')['cylinders'].transform(lambda x: x.fillna(x.median()))
df['odometer'] = df.groupby(['model_year', 'model'])['odometer'].transform(lambda x: x.fillna(x.median()))

# Remove outliers
df = df[(df['price'] >= 500) & (df['price'] <= 50000)]  # Example price range
df = df[(df['model_year'] >= 1990) & (df['model_year'] <= 2022)]  # Example model year range




```

```python
# Histograms
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['model_year'], bins=30, kde=True)
plt.title('Distribution of Car Model Years')
plt.xlabel('Model Year')
plt.ylabel('Frequency')
plt.show()

# Scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='odometer', y='price', data=df)
plt.title('Price vs Odometer')
plt.xlabel('Odometer')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='model_year', y='price', data=df)
plt.title('Price vs Model Year')
plt.xlabel('Model Year')
plt.ylabel('Price')
plt.show()


```

## Data Preprocessing
- Filled missing values in `model_year`, `cylinders`, and `odometer`.
- Removed outliers in `price` and `model_year`.

## Visualizations
- Histograms of car prices and model years.
- Scatter plots of price vs odometer and price vs model year.

## Conclusion
This analysis provides insights into the distribution of car prices, model years, and their relationships with odometer readings. The preprocessing steps help in making the data more consistent and the visualizations more informative.
